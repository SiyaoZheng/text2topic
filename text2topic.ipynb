```python
num_topics = 100
num_top_words = 50
```

```python
stopwords = [ line.rstrip().decode('gbk').encode('utf8') for line in open('/home/siyao/stopwords/baidu.txt') ]
for stopword in stopwords:
    print stopword
```

```python
report= pd.read_csv('/home/siyao/lgov(1).csv')
report.head()
```

```python
txt_list = []
cixing=["x","zg","uj","ul","e","d","uz","y"]
for i, txt in enumerate(report['text']):
    result = ''
    print "开始处理第" +str(i) + "个啦啦啦啦"
    try:
        for w in pseg.cut(txt):
                if not str(w.flag)  in cixing:
                    seg=str(w.word.encode('utf-8'))
                    if seg not in stopwords:
                            result+=str(seg)+" "
    finally:
        pass
    txt_list.append(result)
    print "完成这个文档了啦啦啦啦"
            
```

```python
vectorizer = CountVectorizer(min_df=1)
```

```python
dtm = vectorizer.fit_transform(txt_list)
```

```python
vocab = vectorizer.get_feature_names() 
```

```python
clf = decomposition.NMF(n_components=num_topics, random_state=1)
doctopic = clf.fit_transform(dtm)
```

```python
topic_words = []
for topic in clf.components_:
    word_idx = np.argsort(topic)[::-1][0:num_top_words]
    topic_words.append([vocab[i] for i in word_idx])
```

```python
words = pd.DataFrame(topic_words)
words.to_csv('/home/siyao/data/lwords.csv', encoding='gb18030')
```

```python
doctopic = doctopic / np.sum(doctopic, axis=1, keepdims=True)
doctopic = pd.DataFrame(doctopic)
report = pd.concat([report, doctopic], axis=1)
try:
    del report['text']
except KeyError:
    pass
report.to_csv('/home/siyao/data/ltest.csv')
```
